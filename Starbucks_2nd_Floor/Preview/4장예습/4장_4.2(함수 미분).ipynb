### 모형

- **예측 모형의 성능**
    - 분석의 목표는 예측 오차가 가장 적은 최적의 모형 예측이다
    - 예측모형 예(선형예측모형)
$$ \hat y = w_1x_1 +w_2x_2 + ... + w_Nx_N = w^Tx $$
    - w = 가중치(변경 가능)
    - 이중 가장 좋은 모형을 찾는 함수를 **성능함수** 라고 함다
    - 성능함수는 클수록 좋다
    
    - 성능, 손실, 비용, 오차함수들을 목적 함수라 한다

- 성능함수 : 성능이 가장 크게되는 모형을 찾는 함수
- 목적함수 : 성능함수, 손실함수 등 최적화의 대상이 되는 함수
- 최적화 : 목적함수의 값을 가장 크게 또는 가장 작게 되는 점을 찾는것
- **최적화 하기 위해 미분을 사용한다**

- **기울기**
    - 기을기(slope):
$$ slope = \frac{\Delta y}{\Delta x} = \frac{f(x_2) - f(x)}{x_2 - x} = \lim_{\Delta x\to0}\frac{f(x + \Delta x) - f(x)}{\Delta x} $$

##### $x^3 - 3x^2 + x$ 함수 그레프

```python
def f(x):
    return x**3 - 3 * x**2 + x

y = f(x)
plt.plot(x, y)

plt.xlim(-1.5, 3.5)
plt.ylim(-4, 2)
plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-4, 2))

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("함수의 기울기")

plt.show()
```

### 미분

- 수치미분
    - 대략적인 기울기를 구하는 미분
    $$ slope \approx \frac{f(x + \frac{1}{2}dx) - f(x - \frac{1}{2}dx)}{dx} $$

- **미분**
    - **미분** : 어떤 함수로부터 그 함수 기울기를 출력하는 새로운 함수를 만들어내는 작업.
    - **도함수** : 미분으로 만들어진 함수

$$ f' = \frac{d}{dx}(f) = \frac{d}{dx}f = \frac{df}{dx} = \frac{d}{dx}(y) = \frac{d}{dx}y = \frac{dy}{dx} $$

- $f'$는 "f프라임", $\frac{df}{dx}$는 "df 오버 dx" 라 읽는다

- **미분 가능**
    - 한 점에서 기울기 값이 1개가 있어야 한다
    - 만약 기울기를 정의할 수 없을경우 미분 불가능이라 한다
    - 예) ReLU 함수

```python
def relu(x):
    return np.where(x > 0, x, 0)

xx = np.linspace(-2, 2, 100)
plt.plot(xx, relu(xx))
plt.plot(0, 0, marker='o', ms=10)
plt.title("ReLU")
plt.xlabel("$x$")
plt.ylabel("$ReLU(x)$")
plt.show()
```

### 미분공식

- **도함수를 구하는 공식**
     1. 기본 미분공식
     2. 선형조합법칙
     3. 곱셈법칙
     4. 연쇄법칙

- **기본미분 공식**
    - **상수** : 상수를 미분하면 0이다
    $$ \frac{d}{dx}(c) = 0$$
    - **거듭제곱** : x의 n제곱을 미분하면 n-1 제곱으로 변수가 1씩 감소한다, (n 이 자연수이거나 음수일때만)
    $$ \frac{d}{dx}(x^n) = nx^{n - 1}  $$
    - **로그** : 로그함수를 미분하면 $x^{-1}$이 된다
    $$ \frac{d}{dx}(\log x) = \frac{1}{x} $$
    - **지수** : 밑이 오일러 수인 지수함수는 미분해도 변하지 않는다
    $$ \frac{d}{dx}(e^x) = e^x $$

- **선형 조합 법칙**
    - 상수 곱
    $$ \frac{d}{dx}(cf) = c * \frac{d}{dx}(f) $$
    - 함수 덧셈
    $$ \frac{d}{dx}(f_1 + f_2) = \frac{d}{dx}f_1 + \frac{d}{dx}f_2 $$
    - 선형 조합
    $$ \frac{d}{dx}(c_1f_1 + c_2f_2) = c_1\frac{d}{dx}f_1 + c_2\frac{d}{dx}f_2 $$

- **곱셈법칙**
    - 서로 다른 함수의 곱으로 이루어진 함수의 도함수
    $$ \frac{d}{dx}(f*g) = f*\frac{d}{dx}g + g*\frac{d}{dx}f $$

- **연쇄법칙**
    - 미분하고자 하는 함수의 입력변수가 다른 함수의 출력변수인 경우
    $$ f(x) = h(g(x)) $$
    $$ \frac{d}{dx}f = \frac{dh}{dg}\frac{d}{dx}g $$

- **2차 도함수**
    - 도함수를 한 번 더 미분하여 만들어진 함수
  $$ f'' = \frac{d^2}{dx^2}(f) = \frac{d^2}{dx^2}f = \frac{d^2f}{dx^2} = \frac{d^2}{dx^2}(y) = \frac{d^2}{dx^2}y = \frac{d^2y}{dx^2} $$
    이는 도함수의 기울기를 나타낸다

- 2차 도함수 값의 성질
    - 2차 도함수 값이 **양수**이면 **볼록**
    - 2차 도함수 값이 **음수**이면 **오목**
    - 따라서 2차 도함수값을 볼록도라고도 한다

- **편미분**
    - 편미분결과가 각각 해당 기울기이다
    - 독립변수가 둘 이상일 경우 하나의 변수의 기울기를 구하는 미분 
    $$ f_x(x, y) = \frac{\partial}{\partial x}f $$
    $$ f_y(x, y) = \frac{\partial}{\partial y}f $$
    - 이때 해당되지않는 변수는 상수취급

- **다변수 함수의 연쇄법칙**
    - 다변수 함수의 미분을 구할때 함수가 연결되어 있으면 연쇄법칙이 적용된다
        $$ y_1 = f_1(x) $$
        $$ y_2 = f_2(x) $$
        $$ z = g(y1, y2) $$
        이때 함수 g를 x에 대하여 미분하면
        $$ \frac{d}{d x}z = \frac{\partial z}{\partial y_1}\frac{dy_1}{dx} + \frac{\partial z}{\partial y_2}\frac{dy_2}{dx} $$

- **2차 편미분**
    - 편미분의 2차 도함수
        1. 첫 미분 두번째 미분 둘다 x 
        $$ f_{xx}(x, y) = \frac{\partial^2}{\partial x^2}f $$
        2. 첫 미분 두번째 미분 둘다 y
        $$ f_{yy}(x, y) = \frac{\partial^2}{\partial y^2}f $$
        3. 첫 미분은 x 두번째 미분은 y
        $$ f_{xy}(x, y) = \frac{\partial^2}{\partial y\partial x}f $$
        4. 첫 미분은 y 두번째 미분은 x
        $$ f_{yx}(x, y) = \frac{\partial^2}{\partial x\partial y}f $$
- 이때 $ f_{yx}(x, y) = f_{xy}(x, y) $이다 이를 슈와르츠 정리라고 한다

- **테일러 전개**
    - 함수의 기울기를 알고 있다면 함수의 모양을 다음처럼 근사화 할 수 있다.
    $$ f(x)  \approx f(x_0) + \frac{df(x_0)}{dx}(x - x_0) $$
    - 다변수 함수
    $$ f(x, y)  \approx f(x_0, y_0) + \frac{df(x_0,y_0)}{dx}(x - x_0) + \frac{df(x_0,y_0)}{dy}(y - y_0) $$

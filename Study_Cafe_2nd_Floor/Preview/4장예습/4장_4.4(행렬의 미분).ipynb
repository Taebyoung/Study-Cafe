# 행렬의 미분

- 벡터와 행렬을 함수로 표현
    1. 벡터
    $$ f\left(\begin{bmatrix}x_1 \\x_2 \end{bmatrix} \right) = f(x) = f(x_1, x_2) $$
        - vector x $\to$ scalar f
    2. 행렬
    $$ f\left(\begin{bmatrix}x_{11} & x_{12} \\ x_{21} & x_{22} \end{bmatrix} \right) = f(X) = f(x_{11},\cdots,x_{22}) $$
        - matrix x $\to$ scalar f
    3. 스칼라를 벡터로
    $$ f(x) = \begin{bmatrix}f_1(x)\\f_2(x)\end{bmatrix} $$
        - scalar x $\to$ vector f
    4. 스칼라를 행렬로도 가능
    $$ f(x) = \begin{bmatrix}f_{11}(x) & f_{12}(x) \\ f_{21}(x) & f_{22}(x)\end{bmatrix} $$
        - scalar x $\to$ matrix f

***
- **스칼라를 벡터로 미분하는 경우**
    - **그레디언트 벡터**
$$ \nabla f = 
\frac{\partial f}{\partial {x}} =
\begin{bmatrix}
\dfrac{\partial f}{\partial x_1}\\
\dfrac{\partial f}{\partial x_2}\\
\vdots\\
\dfrac{\partial f}{\partial x_N}\\
\end{bmatrix} $$
    - 이렇게 만들어진 벡터를 **그레디언트 벡터** 라고 하고 $\nabla f$로 표기한다
- 예)
$$ f(x, y) = 2x^2 + 6xy + 7y^2 - 26x - 54y + 107 $$
$$ \nabla f = 
\begin{bmatrix}
\dfrac{\partial f}{\partial x}\\
\dfrac{\partial f}{\partial y}\\
\end{bmatrix} =
\begin{bmatrix}
4x + 6y - 26\\
6x + 14y - 54\\
\end{bmatrix} $$

    1. $$ f(x, y, z) = x + y + z $$
    2. $$ f(x, y, z) = xyz $$

## 퀴즈 5 
그레디언트 벡터 
1. 
$$ \nabla f = \begin{bmatrix}4x + 6y - 26\\6x + 14y - 54\\ \end{bmatrix} $$

에 대해서 $x, y$가 다음 위치일 때 그레디언트 벡터의 값을 구하고 평면상에 화살표로 나타내라. 
1. $x=7, y=1$
2. $x=2, y=1$

- **컨투어플롯, 퀴버플롯**
    - 2차원 함수 f(x,y)는 평면상에서 컨투어 플롯으로 나타낼 수 있다
    - 입력변수 x,y 위치에서 그레디언트 벡터값은 그 위치를 원점으로 하는 화살표로 표현할 수 있다.
    - 이렇게 컨투어 플롯 위에 그레디언트 벡터를 화살표로 나타낸 것을 퀴버플롯 이라 한다

```python
def f(x, y):
    return 2 * x**2 + 6 * x * y + 7 * y**2 - 26 * x - 54 * y + 107

xx = np.linspace(1, 16, 100)
yy = np.linspace(-3, 6, 90)
X, Y = np.meshgrid(xx, yy)
Z = f(X, Y)


def gx(x, y):
    return 4 * x + 6 * y - 26

def gy(x, y):
    return 6 * x + 14 * y - 54

xx2 = [7, 2]
yy2 = [1]
X2, Y2 = np.meshgrid(xx2, yy2)
GX = gx(X2, Y2)
GY = gy(X2, Y2)
plt.figure(figsize=(10, 5))
plt.contour(X, Y, Z, levels=np.logspace(0, 3, 10))
plt.quiver(X2, Y2, GX, GY, color='blue', scale=400, minshaft=2)

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 행렬미분법칙

- 행렬미분법칙 1: **선형모형**
    - 선형 모형을 미분하면 그레디언트 벡터는 가중치 벡터이다 
    $$ f(x) = w^T x $$
    $$ \nabla f = \frac{\partial {w}^{T}{x}}{\partial {x}} = \frac{\partial {x}^{T}{w}}{\partial {x}} = {w} $$ 

- 행렬미분법칙 2: **이차 형식**
    - 이차 형식을 미분하면 행렬과 벡터의 곱으로 나타난다
    $$ f(x) = x^TAx $$
    $$ \nabla f(x) = \frac{\partial x^TAx}{\partial x} = (A + A^T)x $$

***
스칼라 미분   벡터/행렬미분
***
 $ax  \to  a$ $\;\;\;\;\;\;\;$ $w^Tx \to w$              
*** 
 $ax^2 \to 2ax \;\;\;\;\  x^TAx \to (A+A^T)x $      
***

    - 백터를 스칼라로 미분하는 경우
$$ {f}(x) = \begin{bmatrix} f_1 \\ f_2 \\ \vdots\\ f_M \\ \end{bmatrix} $$
$$ \frac{\partial {f}}{\partial x} = \begin{bmatrix} \dfrac{\partial f_1}{\partial x} & \dfrac{\partial f_2}{\partial x} & \cdots & \dfrac{\partial f_M}{\partial x} \end{bmatrix} $$    
    - 벡터를 벡터로 미분하는 경우
        - 벡터를 벡터로 미분하면 미분을 당하는 벡터의 원소가 여러개이고 미분을 하는 원소도 여러개이므로 미분의 결과로 나온 도함수는 행렬이 된다

- 행렬미분법칙 3: 행렬과 벡터의 곱의 미분
    - Ax를 벡터 x로 미분하면 $A^T$이다
$$ f(x) = Ax $$
$$ \nabla f(x) = \frac{\partial(Ax)}{\partial x} = A^T $$
    - 이렇게 만들어진 도함수의 행렬을 **자코비안 행렬** 이라고 한다
    - 그레디언트 벡터의 도함수의 행렬을 **헤시안 행렬**이다
        - $$ Hf(x) = H = J(\nabla f(x))^T $$

- 행렬미분법칙 4: 행렬 곱의 대각성분
    - 두 정방행렬의 곱의 대각성분은 스칼라 이다. 이 스칼라를 뒤 행렬로 미분하면 앞의 전치행렬이 나온다
    $$ f(X) = tr(WX) $$
    $$ \frac{\partial}{\partial X}f = \frac{\partial}{\partial X}tr(WX) = W^T $$

- 헹랼미분법칙 5: 행렬식의 로그
    - 행렬식은 스칼라값이고 이 값의 로그 값도 스칼라이다. 이 값을 원래의 행렬로 미분하면 원래 행렬의 역행렬의 전치행렬이 된다
    $$ f(X) = \log |x| $$
    $$ \frac{\partial}{\partial X}f = \frac{\partial}{\partial X}\log |X| = (X^{-1})^T $$

### 부정적분
    - 정확하게 미분과 반대되는 개념
    - 도함수가 미분되기 전의 원래함수를 찾아가는 과정
$ \int $ (integral)

$$ \frac{d}{dx}F(x) = f(x) \;\leftrightarrow\; F(x) =  \int f(x)dx + c $$

## 퀴즈 4
- 다음 부정적분을 구하라
    1. $ 4x^2 - 2\frac{1}{x^2} + 1 $ 
    2. $ \frac{1}{xln5} $
    3. $ 3^xln3$

- **편미분의 부정적분**
    - 편미분을 한 도함수에서 원래함수를 찾는 방법
        1. x로 편미분하였을 경우
$$ \frac{\partial F_1(x, y)}{dx} = f(x, y) \leftrightarrow F_1(x, y) = \int f(x, y)dx + C(y) $$ 이때 C(y)에는 x가 있으면 안된다
        2. y로 편미분하였을 경우
$$ \frac{\partial F_2(x, y)}{\partial y} = f(x, y) \leftrightarrow F_2(x, y) = \int f(x, y)dx + C(x) $$ 이때 C(x)에는 y가 있으면 안된다

- **다차 도함수와 다중적분**
    - 미분을 여러번 한 도함수를 원래 함수를 찾는 방법
    - 예) $F_3(x,y)$를 x로 미분하고 y로 미분한 2차 도함수
    $$ \frac{\partial^2F_3(x,y)}{\partial x\partial y} = f(x,y) \leftrightarrow F_3(x,y) = \int_x\int_y f(x,y)dydx$$
    적분기호 아래 변수명 생략 가능
    $$ \iint f(x,y)dydx $$

## 정적분
    - 독립변수 x가 어떤 구간[a, b]사이일때 그 구간의 함수 f(x)의 값고 수평선이 이루는 면적을 구하는 행위
$$ \int_a^bf(x)dx$$

- 미적분학의 기본 정리
$$ \int_a^bf(x)dx = F(b) = F(a)$$

```python
from matplotlib.patches import Polygon
def f(x):
    return x ** 3 - 3 * x ** 2 + x + 6
a, b = 0, 2
x = np.linspace(a - 0.5, b + 0.5, 50)
y = f(x)
ax = plt.subplot(111)
plt.title("정적분의 예")
plt.plot(x, y, 'r', linewidth=2)
plt.ylim(bottom=0)
ix = np.linspace(a, b)
iy = f(ix)
verts = [(a, 0)] + list(zip(ix, iy)) + [(b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.5 * (a + b), 0.2 * (f(a) + f(b)), r"$\int_a^b f(x)dx$",
         horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.05, '$x$')
plt.figtext(0.1, 0.9, '$y$')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([])
ax.set_xlim(-2, 4)
ax.set_ylim(0, 8)
plt.show()
```

- 이는 기본정리를 사용하여 푸는 방법과 잘게 쪼개어 면적을 근사하게 구하는 수치적분이 있다

- 수치적분
    - 함수를 아주 작은 구간으로 나누어 실제 면적을 계산함으로써 정적분의 값을 구하는 방법이다

- 다변수 정적분
    - 입력변수가 2개인 2차원 함수 $f(x, y)$의 적분
    $$ \int_{y=c}^{y=d}\int_{x=a}^{x=b}f(x,y)dxdy $$

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
_x = np.arange(12) / 2 + 2
_y = np.arange(12) / 2
X, Y = np.meshgrid(_x, _y)
x, y = X.ravel(), Y.ravel()
z = x * x - 10 * x + y + 50
z0 = np.zeros_like(z)
ax.bar3d(x, y, z0, 0.48, 0.48, z)
ax.set_xlim(0, 10)
ax.set_ylim(-2, 10)
ax.set_zlim(0, 50)
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.title("f(x, y)")
plt.show()
```

- 다차원 함수의 단일 정적분
    - 2차원 함수이지만 이중적분을 하지 않고 단일 정적분을 하는 경우. 이 때는 하나의 변수만 진짜 변수로 보고 나머지 하나는 상수라 간주한다
    $$ \int_a^bf(x,y)\mathrm{d}x $$
